### 讀取 `pkgo_nyc_5class.csv.bz2`

```python
pkgo = pd.read_csv('pkgo_nyc_5class.csv.bz2')
```

```python
''' Import keras to build a DL model '''
from keras.models import Model, Sequential
from keras.layers import Dense, Activation, LeakyReLU
from keras.optimizers import SGD

def build_model(X, acti):
    
    if acti == 'leaky-relu':
        lrelu = LeakyReLU(alpha = 0.02)
    
    model = Sequential()
    model.add(Dense(128, input_dim=X.shape[-1]))
    
    # add activation function
    if acti == 'leaky-relu':
        model.add(lrelu)
    else:
        model.add(Activation(acti))
    
    model.add(Dense(256))
    
    # add activation function
    if acti == 'leaky-relu':
        model.add(lrelu)
    else:
        model.add(Activation(acti))

    model.add(Dense(5))
    model.add(Activation('softmax'))
    return model
```

```python
def Xy_split(pkgo):
    num_cols = pkgo.shape[1]
    X = pkgo.iloc[:, :num_cols-1]
    y = pkgo['class']
    
    # 型別轉換，確保資料型態正確
    X = X.astype(np.float32)
    y = y.astype(np.int)
    
    # 對 y 做 one-hot encoding
    y = to_categorical(y)
    
    # 打亂資料集
    X, y = shuffle(X, y)
    
    return X, y

X, y = Xy_split(pkgo)
```

## 比較不同 activation function 對訓練結果之差異

```python
acti_list = ['sigmoid', 'relu', 'leaky-relu']

import keras.backend as K

batch_size = 32
epochs = 100
all_loss, all_acc = [], []

for acti in acti_list:
    K.clear_session()
    print("Running model, activation function = {}".format(acti))

    model = build_model(X, acti)

    optim = SGD(lr=0.01, momentum=0.0, decay=0.0, nesterov=False)

    model.compile(loss='categorical_crossentropy',
          optimizer=optim,
          metrics=['accuracy'])

    history = model.fit(X, y,
                batch_size=batch_size,
                epochs=epochs,
                verbose=0,
                shuffle=True,
                validation_split=0.1)

    loss = history.history.get('loss')
    acc = history.history.get('acc')

    all_loss.append(loss)
    all_acc.append(acc)
```

```python
''' Visualize the loss and accuracy of different learning rates'''
rcParams['figure.figsize'] = 10, 7

plt.figure(0)
plt.subplot(121)
for k in range(len(acti_list)):
    plt.plot(range(len(all_loss[k])), all_loss[k],label=acti_list[k])
plt.title('Loss')

plt.subplot(122)
for k in range(len(acti_list)):
    plt.plot(range(len(all_acc[k])), all_acc[k],label=acti_list[k])
plt.title('Accuracy')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
